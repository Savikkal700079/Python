# **Metacharacters**

Metacharacters are characters that are interpreted in a special way by a RegEx engine.



1.   **[ ] - Square brackets**
      It specifies a set of characters you wish to match.
2.   **. - Period**
      A period matches any single character (except newline '\n').
3.   **^ - Caret**
      The caret symbol ^ is used to check if a string starts with a certain character.
4.   **$ - Dollar**
      The dollar symbol $ is used to check if a string ends with a certain character.
5.   ***- Star**
      The star symbol * matches zero or more occurrences of the pattern left to it.
6.   **+ - Plus**
      The plus symbol + matches one or more occurrences of the pattern left to it.
7.   **? - Question Mark**
      The question mark symbol ? matches zero or one occurrence of the pattern left to it.
8.   **{} - Braces**
      Consider this code: {n,m}. This means at least n, and at most m repetitions of the pattern left to it. It indicate the number of occurrences of a preceding regex to match.
9.   **| - Alternation**
      Vertical bar | is used for alternation (or operator).
10.  **() - Group**
      Parentheses () is used to group sub-patterns.
11.  **\ - Backslash**
      Backlash \ is used to escape various characters including all metacharacters.










```python
import re
s1="P45on"
p1="^P....n$"
# re.match(pattern,string)
res=re.match(p1,s1)
print(res) # returns a match object
if res:
  print("Match found")
else:
  print("Not found")
```

# **Special sequence**

\A	Returns a match if the specified characters are at the beginning of the string

\b	Returns a match where the specified characters are at the beginning or at the end of a word

\B	Returns a match where the specified characters are present, but NOT at the beginning (or at the end) of a word

\d	Returns a match where the string contains digits (numbers from 0-9)

\D	Returns a match where the string DOES NOT contain digits

\s	Returns a match where the string contains a white space character

\S	Returns a match where the string DOES NOT contain a white space character

\w	Returns a match where the string contains any word characters (characters from a to Z, digits from 0-9, and the underscore _ character)

\W	Returns a match where the string DOES NOT contain any word characters

\Z	Returns a match if the specified characters are at the end of the string

```python
# "\A" Returns a match if the specified characters are at the beginning of the string
import re
s1="Welcome to Python class"
res1=re.findall("\AWe",s1)
print(res1)
res2=re.findall("\AHe",s1)
print(res2)
```

```python
# "\b" Returns a match where the specified characters are at the beginning or at the end of a word
import re
s1="Welcome to Python class"
res1=re.findall(r"\bWe",s1) # checking whether the string starts with We
print(res1)
res2=re.findall(r"ss\b",s1) # checking whether the string ends with ss
print(res2)
```

```python
# "\B" Returns a match where the specified characters are present, but NOT at the beginning (or at the end) of a word
import re
s1="Welcome to Python class"
res1=re.findall(r"\BWe",s1) # checking whether the string "We" is present in the string s1, not at the beginning
print(res1)
res2=re.findall(r"ss\B",s1) # checking whether the string "ss" is present in the string s1, not at the end
print(res2)
res3=re.findall(r"\Bre","Hi, we are learning regular expression") # checking whether the string "re" is present in the string s1, not at the beginning
print(res3)
res4=re.findall(r"or\B","hello world") # checking whether the string "or" is present in the string s1, not at the end
print(res4)
```

```python
# "\d" Returns a match where the string contains digits (numbers from 0-9)
import re
s1="We3lcome to P1ython class"
res1=re.findall("\d",s1)
print(res1)
res2=re.findall("\d","Hello world") # no digits in this string
print(res2)
```

```python
# "\D"	Returns a match where the string DOES NOT contain digits
import re
res1=re.findall("\D","He2llo wor4ld")
print(res1)
res2=re.findall("\D","He2llo wor4ld@")
print(res2)
```

```python
# "\s"	Returns a match where the string contains a white space character
import re
res1=re.findall("\s","Hello all, welcome to python lab") # we have 5 spaces in this string
print(res1)
```

```python
# "\S"	Returns a match where the string DOES NOT contain a white space character
import re
res1=re.findall("\S","Hello all!!") # except spaces remaining characters it will consider
print(res1)
```

```python
# "\w"	Returns a match where the string contains any word characters (characters from a to Z, digits from 0-9, and the underscore _ character)
import re
res1=re.findall("\w","Hel@lo all!12_3") # it takes only letters, digits and underscore --> word characters
print(res1)
```

```python
# "\W"	Returns a match where the string DOES NOT contain any word characters
import re
res1=re.findall("\W","Hel@lo all!123") # except word characters, remaining characters it will take (including space)
print(res1)
```

```python
# "\Z"	Returns a match if the specified characters are at the end of the string
import re
# res1=re.findall("\Zworld","Hello world")
# print(res1)
res2=re.findall("rld\Z","Hello world")
print(res2)
```

# **RegEx functions**

The **re** module provides a set of functions that allows us to search a string for a match

The following are some of the functions:


1.   match(pattern, string, flags)
2.   search(patter, string, flags)
3.   findall(pattern, string)
4.   sub(pattern, replacing character, string, max=0)
5.   split(pattern, string)







```python
# match() function
import re
s1="python"
p1="^[Pp]....n$" # string starting with 'P' or 'p'
# re.match(pattern,string)
res=re.match(p1,s1)
print(res) # returns a match object
if res:
  print("Match found")
else:
  print("Not found")
```

```python
# search() function
import re
s1="Welcome to Python lab4"
a=re.search("^Wel.*4$",s1) # starting with Wel and ending with 4, in between we can have zero or more characters
print(a)
if a:
  print("Match found")
else:
  print("Not found")
```

```python
import re
s1="Welcome to Python lab4"
a=re.search("^Wel.*4$",s1)
print(a.group())
b=re.search("on","let's learn Python")
print(b.group()) # returns the part of the string where there was a match
```

```python
# sub() function
import re
s1="Welcome to Computational Statistics1 ab354"
print(re.sub("a","z",s1))
```
